# 2 - Ex 2 graphs
Our data takes to long to run, so we save all the data and create the graphs afterwards.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("/Users/micha/Documents/GitHub/AcedemicSkills2021/r-norm-210622-153041.npy")
```

```python
target_surface = [0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9, 0.95, 0.99]
rsme_par_10 = np.zeros(len(target_surface))
rsme_non_10 = np.zeros(len(target_surface))
rsme_par_200 = np.zeros(len(target_surface))
rsme_non_200 = np.zeros(len(target_surface))

for i, t in enumerate(target_surface):
    rsme_par_10[i] = data[2][2][0][i][0][0]
    rsme_non_10[i] = data[2][2][0][i][1][0]
    rsme_par_200[i] = data[2][2][3][i][0][0]
    rsme_non_200[i] = data[2][2][3][i][1][0]

plt.plot(target_surface, rsme_par_10)
plt.plot(target_surface, rsme_non_10)
plt.plot(target_surface, rsme_par_200)
plt.plot(target_surface, rsme_non_200)
plt.legend(["Parametric, $n = 10$", "Non-Parametric, $n = 10$", "Parametric, $n = 200$", "Non-Parametric, $n = 200$"])
plt.title("RMSE Parametric vs. Non-Parametric at $n = 10$, $n= 200$")
plt.show()
```

```python
target_surface = [0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 0.9, 0.95, 0.99]
plr_par = np.zeros(len(target_surface))
plr_non = np.zeros(len(target_surface))

for i, t in enumerate(target_surface):
    plr_par[i] = data[2][2][0][i][0][1]
    plr_non[i] = data[2][2][0][i][1][1]

plt.plot(target_surface, plr_par)
plt.plot(target_surface, plr_non)
plt.legend(["Parametric", "Non-Parametric"])
plt.title("PLR Parametric vs. Non-Parametric at $n = 10$")
plt.show()
```
